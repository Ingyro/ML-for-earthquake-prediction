# Procesamiento del Vector de entrenamiento

```python
#Importo las librerias pandas, numpy y datetime
import pandas as pd
import numpy as np
import datetime
```

```python
#Cargo el catálogo declusterizado
df = pd.read_csv("salida2.csv", sep=',', decimal='.')
```

```python
#Ordeno el catálogo por cronología siendo el primer evento el más antigüo y el último el más reciente del catálogo
df['date'] = pd.to_datetime(df[['year', 'month', 'day', 'hour', 'minute']])
df =df.sort_values(by='date',ascending=True)
print(df)
```

# Hallamos el b-value de todo el catálogo

```python
#Cálculamos el b-value para todo el catálogo
magnitud = df.iloc[:,16].tolist()
b_value=[]
for idx, val in enumerate(magnitud,start=0):
    sumatoria=0
    for j in range(0,49):      
        indice = (idx - j)
        sumatoria += (magnitud[indice]-2)
    a=0.4342944819/(0.02*sumatoria)
    b_value.append(a)
print(b_value)
print(len(b_value))
```

# Hallamos los primeros 5 valores del vector de entrenamiento

```python
#Calculamos las primeras 5 características del Vector de entrenamiento.
Tv={}
for i in range(0,len(b_value)):
    x1=b_value[i]-b_value[i-4]
    x2=b_value[i-4]-b_value[i-8]
    x3=b_value[i-8]-b_value[i-12]
    x4=b_value[i-12]-b_value[i-16]
    x5=b_value[i-16]-b_value[i-20]
    Tv[i]={x1,x2,x3,x4,x5}
print(len(Tv))
print(Tv)
```

```python
#Convertirmos el arreglo Tv en un dataframe para visualizarlo con más facilidad
Training = pd.DataFrame()
Training= pd.DataFrame.from_dict(Tv, orient='index',
                       columns=['x1', 'x2', 'x3', 'x4','x5'])
Training
```

# Hallamos x6

```python
#Extraemos la fecha y la magnitud para poder calcular mas fácil la característica x6 del Vector de entrenamiento
data = pd.DataFrame()
data['date'] = df['date']
data['magnitud']=df['magnitude']
```

```python
#Visualizamos el dataframe generado
df=data.set_index(['date'])
df
```

```python
#Calculamos el x6
w=pd.DataFrame()
j=[]
inicial=df.index[548]
final=df.index[548]-datetime.timedelta(days=7)
df[final:inicial]
for i in range(len(data)):
    inicial=df.index[i]
    final=df.index[i]-datetime.timedelta(days=7)
    w=df[final:inicial].max()
    j.append(w)
x6 = np.array(j)
print(len(x6))
print(x6)
```

```python
#Agregamos la lista anterior al vector de entrenamiento
Training['x6']=x6
Training
```
